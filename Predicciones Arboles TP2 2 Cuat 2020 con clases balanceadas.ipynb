```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.legend import Legend
import matplotlib.patches as mpatches
import seaborn as sns
#from statsmodels.graphics.mosaicplot import mosaic
import io
import requests

%matplotlib inline


from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.metrics import log_loss

pd.set_option('display.max_columns', None)

```

```python
url_train="https://raw.githubusercontent.com/joacols/TP_orga_de_datos/main/Train_Preprocesado_TP2_Datos_2020-2C-encodeado.csv"
s=requests.get(url_train).content
df_train=pd.read_csv(io.StringIO(s.decode('utf-8')))
df_train.head()
```

```python
url_test="https://raw.githubusercontent.com/joacols/TP_orga_de_datos/main/Test_Preprocesado_TP2_Datos_2020-2C-encodeado.csv"
s=requests.get(url_test).content
df_test=pd.read_csv(io.StringIO(s.decode('utf-8')))
df_test.head()
```

## Separamos la variable a predecir

```python
X = df_train.drop(['Stage','Sales_Contract_No'], axis=1)
```

```python
y = df_train['Stage']
```

```python
X.head()
```

```python
y.head()
```

## Creamos set de entrenamiento y test


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
X_train.head()
```

```python
X_test.head()
```

## DecisionTreeClassifier 

Realizando predicciones con un DecisionTreeClassifier

```python
from sklearn import tree
```

```python
clfDT = tree.DecisionTreeClassifier(class_weight="balanced")
```

```python
clfDT.fit(X_train, y_train)
```

```python
y_pred =clfDT.predict(X_test)
```

```python
#valor previo sin balanceo de clases (como referencia) 0.7734890807516506
accuracy_score(y_test, y_pred)
```

```python
y_pred_probs = clfDT.predict_proba(X_test)
```

```python
#valor previo sin balanceo de clases (como referencia) 7.823409990924415
log_loss(y_test, y_pred_probs)
```

## DecisionTreeClassifier con max_depth = 5

```python
clfDT = tree.DecisionTreeClassifier(max_depth=5,class_weight="balanced")
```

```python
clfDT.fit(X_train, y_train)
```

```python
y_pred =clfDT.predict(X_test)
```

```python
#valor previo sin balanceo de clases (como referencia) 0.7633316404266125
accuracy_score(y_test, y_pred)
```

```python
y_pred_probs = clfDT.predict_proba(X_test)
```

```python
#valor previo sin balanceo de clases (como referencia) 0.5777679375121131
log_loss(y_test, y_pred_probs)
```

## DecisionTreeClassifier con max_depth = 5 Generando CSV para Kaggle

```python
y_pred_probs =clfDT.predict_proba(df_test.drop(['Sales_Contract_No','Stage'], axis=1))[:,1]
```

```python
#Creo un dataframe con el formato ( id , precio (El precio predecido ))
prediccion = df_test['Opportunity_ID'].to_frame()
prediccion.insert(1,'Target', y_pred_probs , True) 
#Exportamos la prediccion en formato csv sin labels (ya subido)
prediccion.to_csv(r'./DecisionTree_max_depth_5_class_weight=balanced.csv',index=False , header=True)
```
